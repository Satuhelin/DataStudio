```python
df4=df3.sort_values(by='country')
```

```python
df_p = pd.read_csv('population - Sheet1.csv')
```

```python
df_p.columns=['country', 'population']
```

```python
df_p=df_p.sort_values(by='country')
```

```python
df_p.dtypes
```

```python
df_all=pd.merge(df_p, df4, left_on='country', right_on='country')
```

```python
df4.dtypes
```

```python
df_all
```

```python
df_all.dtypes
```

```python
df_all.population=df_all.population.str.replace(',', '')
```

```python
df_all.dtypes
```

```python
df_all['population'] = df_all['population'].astype(int)
```

```python
df_all.dtypes
```

```python
df_all['per_person'] = df_all.population / df_all.total
```

```python
pd.options.display.float_format = '{:10}'.format
```

```python
df_most=df_all.sort_values(by='per_person').head(10)
```

```python
df_most
```

```python
df_most.sort_values(by='per_person', ascending=False).head(10).plot(y='per_person', figsize=(8,6), x='country', kind='barh', title='Least people per one rollercoaster')

plt.savefig("most_rollercaosters.pdf")
```
