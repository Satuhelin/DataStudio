```python
profs = list(df2.Ammattiryhmä.unique())
```

```python
profs
```

```python
colors = ['red', 'blue', 'black', 'yellow', 'pink', 'lightgrey','green', 'purple', 'orange', 'lightgreen', 'lightblue', 'dodgerblue']
num=0

fig, ax=plt.subplots()

for prof in profs:
    col = colors[num]
    filt = df2[df2.Ammattiryhmä==prof]
    filt.plot(x='vuosi', y='työntekijöitä', kind="line",ax=ax, color=col, legend='työntekijöitä',
              ylim=(0,3000) ,
              yticks=(np.arange(0,3000, step=100)),
             figsize=(11,13))
    num+=1



plt.savefig("viivat2.pdf")
plt.show()
```
